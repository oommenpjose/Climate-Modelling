```python
with open(r"SData_Out.txt") as datFile:
    p=[data.split() for data in datFile]
# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
x=[]
y=[]
z=[]
for i in range(len(p)):
    for j in range(len(p[0])):
        x.append(j)
        y.append(i)
        z.append(float(p[i][j]))
# Creating figure
fig = plt.figure(figsize =(14, 9))
ax = plt.axes(projection ='3d')
plt.xlabel("Time")
plt.ylabel("Distance")
# Creating plot
ax.scatter3D(y, x, z, color = "green")
plt.title("simple 3D scatter plot")
plt.savefig("THC.png")
# show plot
plt.show()
```
